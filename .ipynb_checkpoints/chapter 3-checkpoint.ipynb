```python
from scipy.io import loadmat
from matplotlib import pyplot as plt
import numpy as np
import math
import scipy
```

```python
x = loadmat('C:/Users/rajra/OneDrive/Desktop/books/analysis of neural data/Ch3-EEG-2')
```

```python
time = x['t']
t = time[0,:]
eeg = x['EEG']
eeg = eeg[:,0]
```

```python
plt.xlim(0, 1)
plt.ylim(-1.5, 1.5)
plt.plot(t,eeg)
```

```python
mn = np.mean(eeg)
```

```python
lags,ac,j,k = plt.acorr(eeg-mn, maxlags = 100,normed = True)
```

```python
del_t = t[1] - t[0]
plt.plot(lags*del_t,ac)
```

```python
1/((lags[17]-lags[0])*del_t)
```

```python
xf = np.fft.fft(eeg-mn)
Sxx = (2*del_t*del_t*xf*np.conj(xf))/(len(eeg)*del_t)
N = len(eeg)/2
Sxx = Sxx[0:int(N)]
```

```python
df = 1/(len(eeg)*del_t)
sampling_f = 1/del_t
nf = sampling_f/2
```

```python
faxis = np.arange(0.0, nf, df)
```

```python
plt.plot(faxis,Sxx)
plt.xlim(0, 100)
```

```python
eeg_60Hz_modeled = []*len(t)
for i in range(len(t)):
    eeg_60Hz_modeled.append(0 + 0.9989*math.sin(2*pi*60*t[i]) - 0.0032*math.cos(2*pi*60*t[i]))
```

```python
from sklearn import preprocessing
eeg_n = eeg.reshape(len(eeg), 1)
scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1))
scaler.fit(eeg_n)
eeg_scaled= scaler.transform(eeg_n)
                      
```

```python
eeg_scaled = np.squeeze(eeg_scaled)
eeg_clean = eeg_60Hz_modeled - eeg_scaled
plt.plot(t,eeg_clean)
plt.ylim(-1.5, 1.5)
```

```python
plt.ylim(-100, 0)
plt.xlim(0, 100)
plt.plot(faxis,10*np.log10(Sxx/max(Sxx)))
```

```python
Fs = 1/del_t
Pxx, freqs, bins, im = plt.specgram(eeg-mn,Fs=Fs,NFFT=1000,noverlap=950, cmap = 'plasma')
plt.ylim(0, 100)
plt.show()
```
