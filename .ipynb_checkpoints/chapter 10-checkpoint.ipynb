```python
#Planning acf
acf1 = np.zeros((50,1999))
for i in range(50):
    plan = spike_train[i,plan_ind]
    lags,coef1,_,_ = plt.acorr(plan-np.mean(plan, axis =0),maxlags = 999)
    plt.close()
    acf1[i,:] = coef1
plt.stem(np.arange(0,100,1),np.mean(acf1[:,1001:1101], axis = 0),use_line_collection = True)
```

```python
#Moving acf
acf2 = np.zeros((50,1999))
for i in range(50):
    plan = spike_train[i,move_ind]
    lags,coef2,_,_ = plt.acorr(plan-np.mean(plan, axis =0),maxlags = 999)
    plt.close()
    acf2[i,:] = coef2
plt.stem(np.arange(0,100,1),np.mean(acf2[:,1001:1101], axis = 0),use_line_collection = True)
```

```python
plan = spike_train[i,plan_ind]
del_t = t[1] - t[0]
T = len(plan)*del_t

xf = np.fft.fft(plan-np.mean(plan))
Sxx = (2*del_t*del_t*xf*np.conj(xf))/T
N = len(plan)/2
Sxx = Sxx[0:int(N)]

df = 1/(T)
sampling_f = 1/del_t
nf = sampling_f/2
faxis = np.arange(0.0, nf, df)
plt.plot(faxis,10*np.log10(Sxx/max(Sxx)))
plt.xlim(0,0.5)
```

```python
#GLM model 1
k = 50
T0 = len(t)
bin_dec = [[0]*len(plan_ind) + [1]*len(move_ind) for i in range(50)]
bin_dec = np.stack(bin_dec,axis=0)
r_bin_dec = np.reshape(bin_dec,(k*T0))
y = np.reshape(spike_train,(k*T0))
data = {'X':r_bin_dec,'y':y}
df = pd.DataFrame(data)
poisson_results1 = smf.glm(formula='y ~ X',data=df,family=sm.families.Poisson()).fit()

lam1 = np.exp(poisson_results1.params[0] + poisson_results1.params[1]*r_bin_dec)
spike_indices_y = np.squeeze(np.argwhere(y == 1))
n = len(spike_indices_y)
z = [0]*n
z[0] = np.sum(lam1[0:spike_indices_y[0]])
for i in range(1,n):
    z[i] = np.sum(lam1[spike_indices_y[i-1]:spike_indices_y[i]])
ecdf = ed.ECDF(z)
f_ecdf = ecdf(z)
zvals = scipy.stats.zscore(z)
mcdf = 1 - np.exp(-zvals)
mcdf = (mcdf-min(mcdf))/(max(mcdf)-min(mcdf))
plt.plot(mcdf,f_ecdf,'o')
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1)+(1.36/math.sqrt(n)))
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1)-(1.36/math.sqrt(n)))
```

```python
#GLM model 2
direction_v = np.zeros((50,2000))
direction_v[right_ind,:] = np.ones(2000)
df['is_right'] = np.reshape(direction_v,(k*T0))
poisson_results2 = smf.glm(formula='y ~ X + is_right',data=df,family=sm.families.Poisson()).fit()
lam2 = np.exp(poisson_results2.params[0] + poisson_results2.params[1]*r_bin_dec + poisson_results2.params[2]*np.reshape(direction_v,(k*T0)))
spike_indices_y = np.squeeze(np.argwhere(y == 1))
n = len(spike_indices_y)
z = [0]*n
z[0] = np.sum(lam2[0:spike_indices_y[0]])
for i in range(1,n):
    z[i] = np.sum(lam2[spike_indices_y[i-1]:spike_indices_y[i]])
ecdf = ed.ECDF(z)
f_ecdf = ecdf(z)
zvals = scipy.stats.zscore(z)
mcdf = 1 - np.exp(-zvals)
mcdf = (mcdf-min(mcdf))/(max(mcdf)-min(mcdf))
plt.plot(mcdf,f_ecdf,'o')
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1)+(1.36/math.sqrt(n)))
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1)-(1.36/math.sqrt(n)))
```

```python
#GLM model 3
order = 70
T1 = T0 - order
y_3 = np.reshape(spike_train[:,order:],(k*T1))
direction_3 = np.reshape(np.expand_dims(direction, axis = 1)*np.ones((1,T1)), (k*T1)).tolist()
ord_plan = int(T0/2 - order)
ord_move = int(T0/2)
h_spikes = np.array([0]*ord_plan + [1]*ord_move)
bin_dec = np.ones((k,1))*np.expand_dims(h_spikes,axis = 0)
r_bin_dec = np.reshape(bin_dec, (k*T1))

data = {'X':r_bin_dec,'y':y_3, 'direction':direction_3}
df_3 = pd.DataFrame(data)

x_hist = []
end = spike_train.shape[1]
df_col_name0 = "hist_"
for i in range(1,order+1):
    reshaped_train = np.reshape(spike_train[:,order+1-i:end+1-i], (k*T1)).tolist()
    df_col_name = df_col_name0 + str(i)
    df_3[df_col_name] = reshaped_train
df_3 = df_3.drop('hist_1',axis=1)

col_names = df_3.columns.tolist() 
feature_str = ' + '.join(col_names[2:])
features_to_fit = 'X +' + ' '+ feature_str
formula = 'y ~' + features_to_fit
poisson_results3 = smf.glm(formula=formula ,data=df_3,family=sm.families.Poisson()).fit()
sum_0_2 = poisson_results3.params[0] + poisson_results3.params[1]*r_bin_dec + poisson_results3.params[2]*df_3['direction']
sum_3_end = 0
for i in range(3,72):
    sum_3_end = sum_3_end + poisson_results3.params[i]*df_3.iloc[:,i]
lam3 = np.exp(sum_0_2 + sum_3_end)

spike_indices_y = np.squeeze(np.argwhere(y == 1))
n = len(spike_indices_y)
z = [0]*n
z[0] = np.sum(lam3[0:spike_indices_y[0]])
for i in range(1,n):
    z[i] = np.sum(lam3[spike_indices_y[i-1]:spike_indices_y[i]])
ecdf = ed.ECDF(z)
f_ecdf = ecdf(z)
zvals = scipy.stats.zscore(z)
mcdf = 1 - np.exp(-zvals)
mcdf = (mcdf-min(mcdf))/(max(mcdf)-min(mcdf))
plt.plot(mcdf,f_ecdf,'o')
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1)+(1.36/math.sqrt(n)))
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1)-(1.36/math.sqrt(n)))
```
